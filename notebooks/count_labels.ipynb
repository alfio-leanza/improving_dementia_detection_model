```python
# Step 1: Import the required libraries
import pandas as pd

# Step 2: Load the CSV file into a pandas DataFrame
df = pd.read_csv('../data/true_pred.csv')

# Step 3: Inspect the first few rows to ensure it loaded correctly
print("First 5 rows of the DataFrame:")
display(df.head())

# Step 4: Count rows where pred_label == true_label and pred_label != true_label
df_correct = df[df['pred_label'] == df['true_label']]
df_incorrect = df[df['pred_label'] != df['true_label']]

# Step 5: Group by 'dataset' column to count per dataset
correct_counts_per_dataset = df_correct.groupby('dataset').size()
incorrect_counts_per_dataset = df_incorrect.groupby('dataset').size()

# Step 6: Print the result
print("Number of correct predictions per dataset type:")
print(correct_counts_per_dataset)
print("Number of incorrect predictions per dataset type:")
print(incorrect_counts_per_dataset)

# Step 7: See the count for each label within each dataset
print("\nCorrect counts for each label within each dataset:")
correct_counts_per_dataset_label = df_correct.groupby(['dataset', 'true_label']).size()
print(correct_counts_per_dataset_label)
print("Incorrect counts for each label within each dataset:")
incorrect_counts_per_dataset_label = df_incorrect.groupby(['dataset', 'true_label']).size()
print(incorrect_counts_per_dataset_label)
```
